```python
5.Write a Python program to check if a number is prime. Expected Output (for input 7):
7 is a prime number

```

```python
num = int(input("Enter a number: ")) 

for i in range(2, num + 1):
    if num % i == 0:
        break

if num == i:  
    print(num, "is a prime number")
else:
    print(num, "is not prime number") 
```
