```python
# Catagory the people by their age like children, adult, citizen, senior citizen...
```

```python
age=int(input("age:"))
if(age in range(0,18)):
    print("children")
elif(age in range(18,30)):
    print("adult")
elif(age in range(30,60)):
    print("citizen")
else:
    print("senior citizen")
```
